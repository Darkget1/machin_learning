```python
import pandas as pd
data = pd.read_csv('../data/review_data.csv')
```

```python
data.head()
```

```python
data =data.drop(['score'],axis=True)
```

```python
data.head()
```

```python
import re
data['review']=[re.sub('[^A-Za-z0-9가-힣]', ' ', s) for s in data['review']]
data['review']=data['review'].replace('  ','')
```

```python
print(data)
```

```python
from keras.preprocessing.text import text_to_word_sequence
from keras.preprocessing.text import Tokenizer

token = Tokenizer()
token.fit_on_texts(data['review'])
```

```python
print(token.word_index)
```

```python
x = token.texts_to_sequences(data['review'])
print(x)
a=0
for index in x:
    if a<len(index):
        a=len(index)
print(a)
```

```python
from keras.utils import pad_sequences
classes = data['y'].to_numpy()
```

```python
#패딩
from keras.utils import pad_sequences
padded_x = pad_sequences(x)#서로의 다른길이의 데이터를 4로 맞추기
print('\n패딩 결과:\n',padded_x)
padded_x.shape
```

```python
word_size = len(token.word_index)+1
```

```python
from keras.layers import Dense,Flatten
from keras.layers import Embedding
from keras.models import Sequential
model = Sequential()
model.add(Embedding(word_size,8,input_length=84))
model.add(Flatten())
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(padded_x,classes,epochs=100)
print('\n Accuracy : %.4f'%(model.evaluate(padded_x,classes)[1]))
```

```python
text = '추웟어요 고기 외에는 별로에요'
token = Tokenizer()
token.fit_on_texts([text])
y = token.texts_to_sequences([text])
print(y)
```

```python
padded_y = pad_sequences(y,84)
y_predict = model.predict(padded_y)
print(y_predict)
```
