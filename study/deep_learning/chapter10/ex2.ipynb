```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python

x1= [0,0,1,1]
x2= [0,1,0,1]
y=[0,1,1,0]

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array(y)
```

```python
model = Sequential()
model.add(Dense(30,input_dim=2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',
metrics = ['accuracy'])
history = model.fit(X,y,epochs=1000,batch_size=2)
```

```python
print(model.predict(X))
##정답!!
```

```python
import keras
model = keras.models.Sequential()
model.add(keras.layers.Dense(units=2,input_shape=(2,),activation='sigmoid'))
model.add(keras.layers.Dense(units=1,activation='sigmoid'))
model.compile(loss='mean_squared_error',optimizer=keras.optimizers.SGD(learning_rate=0.3))
```

```python
model.summary()
#input_shape 입력 갯수
#units 노드갯수

#param 의미 : in 2개 노드 2개 로 총 w4개 b2개로 계산된값을 의미한다.

#다음것은 인풋2개에서 노드 1개로 w2개 b1개로 총 3개이다.
```
