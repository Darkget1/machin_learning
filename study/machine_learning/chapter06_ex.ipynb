```python
import matplotlib.pyplot as plt
import pandas as pd
kdrama = pd.read_csv('data/Machine_Learning/top100_kdrama.csv')

tvN = kdrama['Network'].str.contains('tvN')
kdrama[tvN].head(5)[['Name','Network']]
```

```python
kdrama['Rating'].max(),kdrama['Rating'].min()
d = kdrama['Rating']>=kdrama['Rating'].max()
kdrama[d][['Name','Rating']]
```

```python
from collections import Counter
cast_list = list()
for casts in kdrama['Cast'].to_list():
    casts = casts.split(',')
    for cast in casts:
        cast_list.append(cast)
cast_df = pd.DataFrame.from_dict(Counter(cast_list),orient = 'index').rename(columns={0:'Appearance'})
cast_df.sort_values('Appearance',ascending=False,inplace=True)
cast_df.head()
```

```python
import matplotlib.pyplot as plt
import  numpy as np
top5_df = cast_df.head()
top5_df
plt.figure(figsize=(25,10))
plt.title('Actors and actresses in the drama Top5')
plt.bar(top5_df.index,top5_df['Appearance'],color='r')
plt.xlabel('Actors and actresses')
plt.ylabel('count')
plt.show()
```
