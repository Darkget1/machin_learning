```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/Machine_Learning/04_data2.csv')
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']
plt.figure(figsize=(15,4))
plt.plot(PetalLength,color='r',marker='o' ,label='PetalLength')
plt.plot(PetalWidth,color='g',marker='o',label='PetalWidth')
plt.plot(SepalLength,color='b',marker='o',label='SepalLength')
plt.plot(SepalWidth,color='y',marker='o',label='SepalWidth')
plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Line Grapth')
plt.legend()
plt.grid()
plt.show()
```

```python
PetalLength = data['PetalLength'].mean()
PetalWidth = data['PetalWidth'].mean()
SepalLength = data['SepalLength'].mean()
SepalWidth = data['SepalWidth'].mean()
plt.figure(figsize=(10,5))

plt.bar('Petal Length',PetalLength)
plt.bar('Petal Width',PetalWidth)
plt.bar('Sepal Length',SepalLength)
plt.bar('Sepal Width',SepalWidth)

plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Bar Graph')
plt.grid()
plt.show()
```

```python
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']
plt.figure(figsize=(10,4))
plt.scatter(SepalLength,SepalWidth)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal Scatter Graph')
plt.grid()
plt.show()
```

```python
virginica =data.loc[data.Species=='virginica',['Species','SepalLength']]
plt.figure(figsize=(10,6))
plt.hist(virginica['SepalLength'],bins=7,label='Sepal Length')
plt.xlabel('Sepal')
plt.ylabel('Length')
plt.title('Iris Virginica Sepal Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']
plt.figure(figsize=(8,5))

plt.boxplot([PetalLength,PetalWidth],labels=['Petal Length','Petal Width'])

plt.xlabel('Petal')
plt.ylabel('Value')
plt.title('Iris Petal Box Plot')
plt.show()
```
