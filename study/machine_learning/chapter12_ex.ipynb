```python
from konlpy.tag import Okt
okt = Okt()
```

```python
text = '위대한 공식이 길게 늘어서 있는' \
       '거대한 시공에' \
       '짧은 문장을 새겨 보곤 해' \
       '두어 가지 마음까지' \
       '영원히 노를 저을 순 없지만' \
       '몇 분짜리 노랠 지을 수 있어서' \
       '수만 광년의 일렁임을 거두어' \
       '지금을 네게 들려 줄거야'
text
```

```python
sentence_tag =okt.pos(text)
sentence_tag
```

```python
sentence_tag = okt.morphs(text)
sentence_tag
```

```python
adj_list = []
for word, tag in sentence_tag:
       if tag in ['Noun','Adjective']:
              adj_list.append(word)
```

```python
import collections
count = collections.Counter(adj_list)
count
```

```python
count.most_common(5)
```

```python
import re
new_df['정규화본문'] = new_df['본문'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ 가-힣 0-9]',' ')
new_df.head()
```

```python
temp_list = []
for sentence in new_df['정규화본문']:
       s_list = okt.pos(sentence)
       for word , tag in s_list:
              if tag in ['Noun','Adjective']:
                     temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(50)
tag
```

```python
stopword = ['중','기자','이','수','위','등','것','있다','날','씨','위해','원','개','입니다','를','제','의','고','명','있는','전','두','관','윤','파','시']
temp_list = []
for sentence in new_df['정규화본문']:
       s_list = okt.pos(sentence)
       for word , tag in s_list:
              if tag in ['Noun','Adjective']:
                     if word in stopword:
                            pass
                     else:
                            temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(50)
tag
```

```python
tag_200 = counts.most_common(200)

font = 'data/Machine_Learning/NanumGothic.otf'
wc = WordCloud(random_state=1234,font_path=font
               ,width=500
               ,height=300
               ,background_color='white')
```

```python
img_wordcloud = wc.generate_from_frequencies(dict(tag_200))
#워드 클라우드 출력하기
plt.figure(figsize=(10,5))
plt.axis('off') #테두리 선없애기
plt.imshow(img_wordcloud)
```
