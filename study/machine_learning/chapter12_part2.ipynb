```python
moon = open('data/Machine_Learning/speech_moon.txt','rt',encoding='UTF8').read()
```

```python
str = ''
lines = moon.readlines()

for line in lines:
       line = line.strip()
       str +=line

moon.close()
str
```

```python
import re
moon = re.sub('[^가-힣]',' ',moon)
moon
```

```python
from konlpy.tag import Hannanum
hannanum = Hannanum()
```

```python
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df_word = pd.DataFrame({'word':nouns})
df_word
```

```python
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
#같은게 있으면 묶고 묶은갯수를 n으로 표기  #line3 = 단어의 갯수 구하는것
df_word = df_word.groupby('word', as_index=False)\
       .agg(n=('word','count'))\
       .sort_values('n',ascending=False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
import seaborn as sns
import matplotlib.font_manager as fm
font = 'data/Machine_learning/NanumGothic.otf'

# plt.rcParams.update({'font.'})
plt.figure(1,figsize=(15,6))
font_name = font_manager.FontProperties(fname=font).get_name()
plt.rc('font',family=font_name)
plt.barh(df_word[:20]['word'],df_word[:20]['n'],color=sns.color_palette('hls',20))
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
#wc만들기
from wordcloud import WordCloud
font = 'data/Machine_Learning/NanumGothic.otf'
wc = WordCloud(random_state=1234,font_path=font
               ,width=400
               ,height=400
               ,background_color='white')
```

```python
#워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)
#워드 클라우드 출력하기
plt.figure(figsize=(10,10))
plt.axis('off') #테두리 선없애기
plt.imshow(img_wordcloud)
```

```python
from PIL import Image
icon = Image.open('data/Machine_Learning/cloud.png')
```

```python
import numpy as np
img = Image.new('RGB',icon.size,(255,255,255))
img.paste(icon,icon)
img = np.array(img)
```

```python
wc = WordCloud(random_state=1234,font_path=font
               ,width=400
               ,height=400
               ,background_color='white'
               ,mask=img)
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)
#워드 클라우드 출력하기
plt.figure(figsize=(10,10))
plt.axis('off') #테두리 선없애기
plt.imshow(img_wordcloud)
```
