```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
dataset = [['치즈','우유','생수']
           ,['생수','수박','우유','맥주']
           ,['수박','생수','맥주']
           ,['생수','우유','맥주']
           ,['치즈','우유','맥주']]

te = TransactionEncoder()
te_arr = te.fit(dataset).transform(dataset)
te_arr
```

```python
df = pd.DataFrame(te_arr,columns=te.columns_)
df
```

```python
itemset = apriori(df,min_support=0.4,use_colnames=True)
itemset.sort_values('support',ascending=False).head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules1 = association_rules(itemset, metric='lift',min_threshold=0.6)
rules1
```
