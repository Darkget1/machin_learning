```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#텐서 플로의 케라스 api에서 필요한 함수를 불러옵니다.
from keras.models import Sequential
from keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
model = Sequential()
```

```python
#출력값 ,입력 변수, 분석 방법에 맞게끔 모델을 설정합니다.
model.add(Dense(1,input_dim=1,activation='linear'))
```

```python
#오차 수정을 위해 경사 하강법(sgd)을, 오차의 정도를 판단하기 위해
#평균 제곱 오차(mse)를 사용합니다.
model.compile(optimizer='sgd',loss='mse')
```

```python
#오차를 최소화 하는 과정을 2000번 반복합니다.
model.fit(x,y,epochs=2000)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')    #예측 결과를 그래프로 나타냅니다.
plt.show()
```

```python
#임의의 시간을 집어넣어 점수를 예측하는 모델을 테스트해 보게습니다.
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다."%(hour,prediction))
```
