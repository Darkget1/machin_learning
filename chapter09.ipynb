```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
seoul = pd.read_excel('data/Machine_Learning/(2010-2020) weather.xlsx')
seoul.head()
```

```python
seoul.info()
```

```python
seoul.describe()
```

```python
seoul.drop('지점',axis=1,inplace=True)
seoul.head()
```

```python
seoul.columns=['날짜','평균기온','최저기온','최고기온']
seoul.columns
```

```python
seoul.isnull().sum()
```

```python
seoul.dropna(subset=['최고기온'],axis=0,inplace=True)
seoul.isnull().sum()
```

```python
seoul['년도']=seoul['날짜'].dt.year
seoul.head()
```

```python
conditions = (seoul['날짜'].dt.month==8)&(seoul['날짜'].dt.day==15)
seoul0815=seoul[conditions]
seoul0815
```

```python
fig = plt.figure(figsize=(15,7))
X = seoul0815[['년도']]
Y = seoul0815[['평균기온']]
plt.xlabel('년도')
plt.ylabel('평균기온')
plt.scatter(X,Y)
plt.show()
```

```python
model = LinearRegression()
X = seoul0815[['년도']]
Y = seoul0815[['평균기온']]
model.fit(X,Y)
result = model.predict([[2022]])
print(result)
```

```python
x = seoul0815['년도']
y = seoul0815['평균기온']
fp1 = np.polyfit(x,y,2)
f1 = np.poly1d(fp1)
fx = np.linspace(2010,2020)
plt.figure(figsize=(15,7))
plt.scatter(x,y)
plt.plot(fx, f1(fx), ls='dashed',lw=3,color='g')
plt.xlabel('년도')
plt.ylabel('평균기온')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import seaborn as sns
plt.figure(figsize=(15,7))

x=seoul[['년도','최저기온','최고기온']]
y=seoul[['평균기온']]

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,test_size=0.3)
model = LinearRegression()
model.fit(x_train,y_train)

plt.plot(model.predict(x_test[:50]), label='predict')
plt.plot(y_test[:50].values.reshape(-1,1), label='real temp')
plt.legend()
```

```python
plt.figure(figsize=(15,7))
y_predict = model.predict(x_test)
plt.scatter(y_test,y_predict,alpha=0.4)
plt.show()
```

```python
print((model.score(x_train,y_train)))
```
