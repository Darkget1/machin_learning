```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('data/Machine_Learning/7장_연습1.xlsx')
df.isnull().sum()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x=df['age'],y=df['value'])
plt.grid(True)
```

```python
df.drop(df[(df['value']>0.9*1e8)&(df['age']>30)].index,inplace=True)
plt.figure(figsize=(10,5))
plt.scatter(x=df['age'],y=df['value'])
plt.grid()
```
