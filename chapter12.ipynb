```python
import pandas as pd
from konlpy.tag import Okt
okt = Okt()
```

```python
sentence_tag = okt.pos(sentence)
print(sentence_tag)
sentence_tag = okt.pos(sentence,join=True)
print(sentence_tag)
```

```python
text = '이승윤 숨고 싶을 땐 다락이 되어 줄거야 죽고 싶을 땐 나락이 되어 줄거야 울고 싶은 만큼' \
       '허송 세월 해 줄거야 진심이 버거울 땐 우리 가면무도회를 열자'
sentence_tag = okt.pos(text)
sentence_tag
```

```python
import collections
adj_list = []
for word,tag in sentence_tag:
       if tag in ['Noun','adjective']:
              adj_list.append(word)
counts = collections.Counter(adj_list)
tag = counts.most_common(2)
print(tag)
```

```python
import matplotlib.pyplot as plt
font_path = 'data/Machine_Learning/NanumGothic.otf'
f = open('data/Machine_Learning/달이참예쁘다고.txt','rt', encoding='UTF8')

str = ''
lines = f.readlines()

for line in lines:
       line = line.strip()
       str +=line

f.close()
str
```

```python
sentence_tag = okt.pos(str)
sentence_tag
```

```python
adj_list = []
for word, tag in sentence_tag:
       if tag in ['Noun','Adjective']:
              adj_list.append(word)
adj_list[:10]
```

```python
counts = collections.Counter(adj_list)
tag = counts.most_common(10)
tag
```

```python
from wordcloud import WordCloud
wc = WordCloud(font_path=font_path,background_color='skyblue',max_font_size=60)
cloud = wc.generate_from_frequencies(dict(tag))
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.grid(False)
plt.imshow(cloud)
```

```python
import pandas as pd
df = pd.read_excel('data/Machine_learning/chap12_실습2.xlsx')
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df['정규화본문']=df['본문'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ 가-힣 0-9]','')
df.head()
```

```python
temp_list = []
for sentence in df['정규화본문']:
       s_list = okt.pos(sentence)
       for word , tag in s_list:
              if tag in ['Noun','Adjective']:
                     temp_list.append(word)
counts = collections.Counter(temp_list)
tag = counts.most_common(50)
tag
```

```python
wc = WordCloud(font_path=font_path,background_color='white',max_font_size=60,scale=2.0)
cloud = wc.generate_from_frequencies(dict(tag))
plt.figure(figsize=(10,8))
plt.grid(False)
plt.imshow(cloud)
```
